```python
def my_nth_root(x, n, tol):
    
    r = x / n
    
    
    while abs(r**n - x) > tol:
        r = ((n - 1) * r + x / (r**(n - 1))) / n
    
    return r


x = 27
n = 3
tol = 1e-6
root = my_nth_root(x, n, tol)
print(root)
```

```python
def my_fixed_point(f, g, tol, max_iter):

    def F(x):
        return f(x) - g(x)
    
    
    a, b = -1e10, 1e10
    
    
    if F(a) * F(b) >= 0:
        return []

    iter_count = 0
    while iter_count < max_iter:
        
        m = (a + b) / 2
        Fm = F(m)
        
        
        if abs(Fm) < tol:
            return m
        
        
        if F(a) * Fm < 0:
            b = m
        else:
            a = m
        
        iter_count += 1
    
    
    return []

def f(x):
    return x**3

def g(x):
    return x + 1

tol = 1e-6
max_iter = 100
root = my_fixed_point(f, g, tol, max_iter)
if root == []:
    print("The method did not converge within the maximum number of iterations.")
else:
    print(f"The fixed point is approximately {root}")
```

```python
def my_pipe_builder(C_ocean, C_land, L, H):
    
    def total_cost(x):
        
        D_ocean = (x**2 + H**2)**0.5
        
        D_land = L - x
        
        return C_ocean * D_ocean + C_land * D_land
    
    
    tol = 1e-6
    a, b = 0, L
    iter_count = 0
    max_iter = 1000
    
    while (b - a) / 2 > tol and iter_count < max_iter:
        c = (a + b) / 2
        cost_a = total_cost(a)
        cost_b = total_cost(b)
        cost_c = total_cost(c)
        
        
        if cost_c < cost_a and cost_c < cost_b:
            if total_cost(c - tol) < total_cost(c + tol):
                b = c
            else:
                a = c
        elif cost_a < cost_b:
            b = c
        else:
            a = c
        
        iter_count += 1
    
    return (a + b) / 2


C_ocean = 5
C_land = 10
L = 100
H = 20

optimal_x = my_pipe_builder(C_ocean, C_land, L, H)
print(f"The optimal x-value that minimizes the total cost of the pipeline is approximately {optimal_x}")
```

```python
def my_lin_interp(x, y, X):
    Y = []
    for xi in X:
        if xi <= x[0]:
            Y.append(y[0])
        elif xi >= x[-1]:
            Y.append(y[-1])
        else:
            for i in range(len(x) - 1):
                if x[i] <= xi <= x[i + 1]:
                    yi = y[i] + (y[i + 1] - y[i]) * (xi - x[i]) / (x[i + 1] - x[i])
                    Y.append(yi)
                    break
    return Y

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
X = [1.5, 2.5, 3.5]
Y = my_lin_interp(x, y, X)
print(f"Linear interpolation results: {Y}")
```

```python
import numpy as np

def my_cubic_spline(x, y, X):
    n = len(x)
    h = [x[i+1] - x[i] for i in range(n-1)]
    
    # Build the system of equations
    A = np.zeros((n, n))
    b = np.zeros(n)
    
    A[0, 0] = 1
    A[-1, -1] = 1
    
    for i in range(1, n-1):
        A[i, i-1] = h[i-1]
        A[i, i] = 2 * (h[i-1] + h[i])
        A[i, i+1] = h[i]
        b[i] = 3 * ((y[i+1] - y[i]) / h[i] - (y[i] - y[i-1]) / h[i-1])
    
    
    M = np.linalg.solve(A, b)
    
    
    def S(xi, i):
        hi = h[i]
        return (
            M[i] * (x[i+1] - xi)**3 / (6 * hi) +
            M[i+1] * (xi - x[i])**3 / (6 * hi) +
            (y[i] - M[i] * hi**2 / 6) * (x[i+1] - xi) / hi +
            (y[i+1] - M[i+1] * hi**2 / 6) * (xi - x[i]) / hi
        )
    
    Y = []
    for xi in X:
        if xi <= x[0]:
            Y.append(y[0])
        elif xi >= x[-1]:
            Y.append(y[-1])
        else:
            for i in range(n - 1):
                if x[i] <= xi <= x[i + 1]:
                    Y.append(S(xi, i))
                    break
    return Y

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
X = [1.5, 2.5, 3.5]
Y = my_cubic_spline(x, y, X)
print(f"Cubic spline interpolation results: {Y}")
```

```python
import numpy as np

def my_quintic_spline(x, y, X):
    n = len(x)
    h = [x[i+1] - x[i] for i in range(n-1)]
    
    
    A = np.zeros((n, n))
    b = np.zeros(n)
    
    
    A[0, 0] = 1
    A[1, 1] = 1
    A[-1, -1] = 1
    A[-2, -2] = 1
    
    for i in range(2, n-2):
        A[i, i-2] = h[i-2]
        A[i, i-1] = 4 * h[i-2] + h[i-1]
        A[i, i] = 4 * (h[i-2] + h[i-1])
        A[i, i+1] = h[i-1]
        A[i, i+2] = h[i-1]
        b[i] = 5 * ((y[i+1] - y[i]) / h[i-1] - (y[i-1] - y[i-2]) / h[i-2])
    
    
    M = np.linalg.solve(A, b)
    
    
    def S(xi, i):
        hi = h[i]
        return (
            M[i] * (x[i+1] - xi)**5 / (120 * hi) +
            M[i+1] * (xi - x[i])**5 / (120 * hi) +
            (y[i] - M[i] * hi**3 / 120) * (x[i+1] - xi) / hi +
            (y[i+1] - M[i+1] * hi**3 / 120) * (xi - x[i]) / hi
        )
    
    Y = []
    for xi in X:
        if xi <= x[0]:
            Y.append(y[0])
        elif xi >= x[-1]:
            Y.append(y[-1])
        else:
            for i in range(n - 1):
                if x[i] <= xi <= x[i+1]:
                    Y.append(S(xi, i))
                    break
    return Y

x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
X = [1.5, 2.5, 3.5]
Y = my_quintic_spline(x, y, X)
print(f"Quintic spline interpolation results: {Y}")
```

```python
import matplotlib.pyplot as plt
from scipy.interpolate import interp1d

def my_interp_plotter(x, y, X, option):
    if option == 'linear':
        f = interp1d(x, y, kind='linear')
    elif option == 'spline':
        f = interp1d(x, y, kind='cubic')
    elif option == 'nearest':
        f = interp1d(x, y, kind='nearest')
    else:
        raise ValueError("Option must be 'linear', 'spline', or 'nearest'")
    
    Y = f(X)
    
    plt.figure()
    plt.plot(x, y, 'ro', label='Data points')
    plt.plot(X, Y, 'b-', label=f'{option.capitalize()} interpolation')
    plt.title(f'Interpolation using {option.capitalize()} method')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.show()


x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
X = np.linspace(1, 5, 100)
my_interp_plotter(x, y, X, 'linear')
my_interp_plotter(x, y, X, 'spline')
my_interp_plotter(x, y, X, 'nearest')
```

```python
import numpy as np

def my_ls_params(f, x, y):
    
    A = np.zeros((len(x), len(f)))
    for i in range(len(x)):
        for j in range(len(f)):
            A[i, j] = f[j](x[i])
    
    
    beta = np.linalg.lstsq(A, y, rcond=None)[0]
    return beta


x = np.array([1, 2, 3, 4])
y = np.array([2, 3, 5, 7])
f = [lambda x: 1, lambda x: x, lambda x: x**2]

beta = my_ls_params(f, x, y)
print(f"Least squares parameters: {beta}")
```

```python
import numpy as np

def my_func_fit(x, y):
    
    log_x = np.log(x)
    log_y = np.log(y)
    
    
    A = np.vstack([log_x, np.ones(len(log_x))]).T
    beta, log_alpha = np.linalg.lstsq(A, log_y, rcond=None)[0]
    
    
    alpha = np.exp(log_alpha)
    return alpha, beta


x = np.array([1, 2, 3, 4])
y = np.array([2, 4, 8, 16])

alpha, beta = my_func_fit(x, y)
print(f"Fitted parameters: alpha = {alpha}, beta = {beta}")
```

```python
def total_error_cubic(x, y, a, b, c, d):
    
    y_hat = a * x**3 + b * x**2 + c * x + d
    
    error = np.sum((y - y_hat)**2)
    return error

def new_point_no_error(a, b, c, d, x, y):
    
    y_new = a * x**3 + b * x**2 + c * x + d
    return x, y_new


x = np.array([1, 2, 3, 4])
y = np.array([2, 5, 10, 17])
a, b, c, d = 1, 1, 1, -1

error = total_error_cubic(x, y, a, b, c, d)
print(f"Total error: {error}")

x_new = 5
x_new, y_new = new_point_no_error(a, b, c, d, x_new, y)
print(f"New point with no additional error: ({x_new}, {y_new})")
```
